### 04 PyGame

PyGame은 파이선 게임용 라이브러리 입니다. ,pygame을 사용하면 윈도를 만들어 자유롭게 그릴 수 있습니다. 마우스나 키보드 입력도 받습니다. 명령도 풍부합니다.

```python
# 윈도를 화면에 표시합니다.

import sys
import pygame
from pygame.locals import QUIT

pygame.init()
SUR=pygame.display.set_mode((400,300))
pygame.display.set_caption("Just Window")

def main():
    while True:
        SUR.fill((255,0,255))

        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()

        pygame.display.update()

if __name__ == '__main__':
    main()
```

```python
# 메인 루프는 전속력으로 실행을 반복합니다.
#어느 정도 빠른 속도로 메인 루푸가 동작하는지를 확인하는 프로그램을 만들어 보겠습니다.
# 윈도를 화면에 표시합니다.

import sys
import pygame
from pygame.locals import QUIT

pygame.init()
SUR=pygame.display.set_mode((400,300))
pygame.display.set_caption("Just Window")

def main():
    sysfont = pygame.font.SysFont(None,36)
    counter = 0
    while True:

        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()

        counter += 1
        SUR.fill((0,0,0))
        counter_image = sysfont.render(f"counter is {counter}",True,(225,225,225))
        SUR.blit(counter_image,(50,50))
        pygame.display.update()

if __name__ == '__main__':
    main()
```

```python
# 직사각형을 그려보자


import sys
import pygame
from pygame.locals import QUIT,Rect

pygame.init()
SURFACE=pygame.display.set_mode((400,300))
pygame.display.set_caption("Just Window")
FPSCLOCK = pygame.time.Clock()

def main():
    
    while True:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()

        SURFACE.fill((255,255,255))
        r=Rect(30,20,60,40)
        pygame.draw.rect(SURFACE,(20,20,20),r,3)
        pygame.draw.rect(SURFACE,(20,20,20),(50,50,100,100))
        pygame.display.update()
        FPSCLOCK.tick(3)

if __name__ == '__main__':
    main()
```

```python
# 선을 그려보자


import sys
import pygame
from pygame.locals import QUIT,Rect

pygame.init()
SURFACE=pygame.display.set_mode((800,800))
pygame.display.set_caption("Just Window")
FPSCLOCK = pygame.time.Clock()

def main():
    
    while True:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()

        SURFACE.fill((255,255,255))
        for xpos in range(0,800,100):
            pygame.draw.line(SURFACE,(100,100,100),(xpos,0),(xpos,800),2)
            pygame.draw.line(SURFACE,(100,100,100),(xpos+3,0),(xpos+3,800),2)
        
        for ypos in range(0,800,50):
            pygame.draw.line(SURFACE,(200,200,10),(0,ypos),(800,ypos))

        pygame.display.update()
        FPSCLOCK.tick(3)

if __name__ == '__main__':
    main()

```

```python
# 여러선을 그려보자  lines


import sys
import pygame
import random
from pygame.locals import QUIT,Rect

pygame.init()
SURFACE=pygame.display.set_mode((800,800))
pygame.display.set_caption("Just Window")
FPSCLOCK = pygame.time.Clock()

def main():
    
    while True:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
        

        SURFACE.fill((255,255,255))
        pointlist=[]
        for _ in range(10):
            xpos=random.randint(0,800)
            ypos=random.randint(0,800)
            pointlist.append((xpos,ypos))
        pygame.draw.lines(SURFACE,(120,random.randint(0,255),120),True,pointlist)

        pygame.display.update()
        FPSCLOCK.tick(3)

if __name__ == '__main__':
    main()

```

- polygon(SURFACE,color,pointlist,width=0)

```python
# 폴리곤 (다각형)


import sys
import pygame
import random
from math import sin,cos,radians
from pygame.locals import QUIT,Rect

pygame.init()
SURFACE=pygame.display.set_mode((800,800))
pygame.display.set_caption("Just Window")
FPSCLOCK = pygame.time.Clock()

def main():
    i=0
    touch255=False
    while True:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()

        SURFACE.fill((255,255,255))
        pointlist0,pointlist1=[],[]
        for theta in range(0,360,72):
            rad=radians(theta)
            pointlist0.append((cos(rad)*100+200,sin(rad)*100 +150+i*2))
            pointlist1.append((cos(rad)*100+500-i,sin(rad)*100 +150))

        pygame.draw.lines(SURFACE,(120,120,120),True,pointlist0)
        pygame.draw.polygon(SURFACE,(120,i,120),pointlist1)
        pygame.display.update()
        FPSCLOCK.tick(30)

        if touch255 == False:
            i += 10
            if i > 244:
                touch255=True
        elif touch255 ==True:
            i -= 10
            if i < 1:
                touch255=False
if __name__ == '__main__':
    main()

```

```python
# 300*575 랜덤을 중복없이 뽑기

import sys
import pygame
import os, random
from pygame.locals import QUIT, Rect, MOUSEBUTTONDOWN
from PIL import Image


pygame.init()
pygame.key.set_repeat(5, 5)
pygame.display.set_caption("Just Window")
FPSCLOCK = pygame.time.Clock()


folder = "F:/개개비/이미지/"   # 원하는 폴더 경로
files = os.listdir(folder)  # 폴더 안 파일 목록
choice = random.choice(files)

# print("랜덤 선택:", choice)
# imagefile =r"choice"

imagefile=rf"F:\개개비\이미지\{choice}"


img = Image.open(imagefile)
size=200
if img.size[0] > 1200:
    W= 1200//size
else:
    W=img.size[0]//size
H=img.size[1]//size
SURFACE = pygame.display.set_mode((W*size*2, H*size))
print(imagefile)

def main():
    
    img = pygame.image.load(imagefile)
    # W, H = 6, 11  # 가로,세로
    images = []
    rects = []  # ← 각 이미지의 Rect 저장
    placed_list = []  # ← 클릭된 이미지 index를 저장하는 리스트
    placed_index = [] 

    # ------ 이미지 잘라서 images[] 에 저장 ------
    for y in range(H):
        for i in range(W):
            image = pygame.Surface((size, size), pygame.SRCALPHA)
            image.blit(img, (0, 0), Rect(i * size, y * size, size, size))
            images.append(image)

    # ------ 인덱스 섞기 ------
    images_index = list(range(W * H))
    random.shuffle(images_index)

    # ------ 섞인 순서대로 화면에 뿌려질 위치의 rect 생성 ------
    for y in range(H):
        for i in range(W):
            pos = (i * size+2, y * size+2)
            rect = Rect(pos[0], pos[1], size, size)
            rects.append(rect)

    running = True

    while running:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()

            elif event.type == MOUSEBUTTONDOWN and event.button == 1:
                mx, my = event.pos

                # ----- 클릭된 이미지가 무엇인지 찾기 -----
                for idx, rect in enumerate(rects):
                    if rect.collidepoint(mx, my):
                        placed_index.append(idx)              # 왼쪽 그리기 제거용
                        original_index = images_index[idx]    # 우측 그리기용
                        placed_list.append(original_index)
                        break

        SURFACE.fill((220, 220, 220))

        # ------ 왼쪽 그림 그리기 ------
        for idx, rect in enumerate(rects):
            if idx in placed_index:               # 이미 클릭된 것은 그리지 않음
                continue
            SURFACE.blit(images[images_index[idx]], rect.topleft)

# ----- 우측에 클릭된 이미지를 strip 원본 좌표대로 배치 -----
        # ------ 우측 그리기 ------
        for k in placed_list:
            px = k % W
            py = k // W
            pygame.draw.line(SURFACE,(255,0,0),(W*size+2,0),(W*size+2,H*size),3)
            SURFACE.blit(images[k], (W*size+10 + px * size, py * size))
            pygame.draw.line(SURFACE,(255,0,0),(W*size+6,0),(W*size+6,H*size),3)

        pygame.display.update()
        FPSCLOCK.tick(30)


if __name__ == '__main__':
    main()
```

> 마우스의 이동
- 마우스의 입력은 MOUSEDOWN  마우스의 이동은 MOUSEMOTION,  마우스의 해제는 MOUSEBUTTONUP입니다. 이러한 이벤트를 사용해서 마우스 궤적을 그립니다.

```python
# 마우스궤적을 그린다.
import sys
import pygame
import random
from pygame.locals import QUIT,MOUSEBUTTONDOWN,MOUSEMOTION,MOUSEBUTTONUP

pygame.init()
SURFACE=pygame.display.set_mode((800,800))
pygame.display.set_caption("Just Window")
FPSCLOCK = pygame.time.Clock()

def main():
    mousepos=[]
    mousedown = False
    while True:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
            elif event.type ==MOUSEBUTTONDOWN:
                mousedown = True
            elif event.type == MOUSEMOTION:
                if mousedown == True:
                    mousepos.append(event.pos)
            elif event.type == MOUSEBUTTONUP:
                mousedown = False
                # mousepos.clear()
        
        SURFACE.fill((255,255,255))
        
        if len(mousepos) > 1:
            pygame.draw.lines(SURFACE,(255,0,0),False,mousepos,True)
        pygame.display.update()
        FPSCLOCK.tick(30)

       

if __name__ == '__main__':
    main()


```

> 키누르기
- 키누르기는 KEYDOWN이벤트로 판별합니다. 어떤 키가 눌렸는지는 이벤트의 key프로퍼티를 보면 알 수 있습니다. 키코드는 pygame.locals에 정의돼 있습니다.

```python
# 키누르기
import sys
import pygame
import random
from pygame.locals import QUIT,KEYDOWN,K_LEFT,K_RIGHT,K_UP,K_DOWN

pygame.init()
pygame.key.set_repeat(5,5)
SURFACE=pygame.display.set_mode((800,800))
pygame.display.set_caption("Just Window")
FPSCLOCK = pygame.time.Clock()

def main():
    logo= pygame.image.load("blog.png")
    pos=[200,150]
    while True:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
            elif event.type ==KEYDOWN:
                if event.key == K_LEFT:
                    pos[0] -=5
                elif event.key == K_RIGHT:
                    pos[0] +=5
                elif event.key == K_UP:
                    pos[1] -=5
                elif event.key == K_DOWN:
                    pos[1] +=5

        pos[0] = pos[0] % 800
        pos[1] = pos[1] % 800

        SURFACE.fill((225,225,225))

        rect= logo.get_rect()
        rect.center = pos
        SURFACE.blit(logo,rect)
        
        
        pygame.display.update()
        FPSCLOCK.tick(30)

       

if __name__ == '__main__':
    main()


```
