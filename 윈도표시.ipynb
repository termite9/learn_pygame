```python
#pos는 중심좌표   radius는 반경

import sys
import pygame
from pygame.locals import QUIT,Rect,KEYDOWN,K_LEFT,K_RIGHT

pygame.init()
pygame.key.set_repeat(5,5)  #거의 즉시 반복된다
SURFACE = pygame.display.set_mode((300,200))
################
FPSCLOCK = pygame.time.Clock()
#################

def main():
    ### main rutin
    strip=pygame.image.load("strip.png")
    images=[]
    for index in range(9):
        image=pygame.Surface((24,24))
        image.blit(strip,(0,0),Rect(index*24,0,24,24))
        images.append(image)
    counter=0
    pos_x=100
    while True:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
            elif event.type ==KEYDOWN:
                if event.key == K_LEFT:
                    pos_x -= 5
                elif event.key == K_RIGHT:
                    pos_x +=5

        SURFACE.fill((0,0,0))
        
        SURFACE.blit(images[counter %2 +0],(50,50))
        SURFACE.blit(images[counter %2 +2],(100,50))
        SURFACE.blit(images[counter %2 +4],(150,50))
        SURFACE.blit(images[counter %2 +6],(200,50))

        counter +=1

        SURFACE.blit(images[8],(pos_x,150))

        pygame.display.update()
        FPSCLOCK.tick(5)

if __name__ =='__main__':
    main()
```

```python
#pos는 중심좌표   radius는 반경

import sys
import pygame
from pygame.locals import QUIT,Rect

pygame.init()
SURFACE = pygame.display.set_mode((800,800))
################
FPSCLOCK = pygame.time.Clock()
#################

def main():
    ### main rutin
    logo=pygame.image.load("blog.jpg")
    theta=0
    while True:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
        
        theta -= 1

        SURFACE.fill((225,225,225))
        
        #로고를 회전하고, 왼쪽위가 (100,30)인 위치에 로고를 그린다.
        new_logo=pygame.transform.rotate(logo,theta)
        # SURFACE.blit(logo,(0,0))
        SURFACE.blit(new_logo,(100,30))

        # pygame.draw.rect(SURFACE,(255,0,255),((110,80),(180,50)))


        pygame.display.update()
        FPSCLOCK.tick(50)

if __name__ =='__main__':
    main()
```

>  폰트작성은 SysFont명령을 사용합니다 반환값은 Font객체 입니다


> pygame.font.SysFont(name,size,bold=False,italic=False) ->Font
-  name은 폰트명으로 기본폰트를 사용하려면 None을 지정한다
- size는 폰트크기

>  render(text,antialias,color,background=None) ->Surface
- text는 그리는 텍스트
- antialias는 안티엘리언스(윤곽을 부드럽게)   표시는 "True"
- color
- background는 생략가능

```python
#pos는 중심좌표   radius는 반경

import sys
import pygame
from pygame.locals import QUIT,Rect

pygame.init()
SURFACE = pygame.display.set_mode((800,800))
################
FPSCLOCK = pygame.time.Clock()
#################

def main():
    ### main rutin
    sysfont=pygame.font.SysFont(None,72)
    message= sysfont.render("Hello Python", True, (0,128,128))
    message_rect=message.get_rect()
    message_rect.center=(200,200)
    while True:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
        

        SURFACE.fill((225,225,225))
        
        
        SURFACE.blit(message,message_rect)



        pygame.display.update()
        FPSCLOCK.tick(150)

if __name__ =='__main__':
    main()
```

```python
#pos는 중심좌표   radius는 반경

import sys
import pygame
from pygame.locals import QUIT,Rect

pygame.init()
SURFACE = pygame.display.set_mode((400,300))
################
FPSCLOCK = pygame.time.Clock()
#################

def main():
    ### main rutin
    sysfont=pygame.font.SysFont(None,72)
    message= sysfont.render("Hello Python", True, (0,128,128))
    message_rect=message.get_rect()
    message_rect.center=(200,200)
    theta=0
    scale=1

    while True:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
        

        SURFACE.fill((225,225,225))
         # theta += 50
        scale = (theta % 360) / 180
        if theta > 0 :
            theta +=5
        else:
            theta -=5
        tmp_msg=pygame.transform.rotozoom(message,theta,scale)
        tmp_rect=tmp_msg.get_rect()
        tmp_rect.center =(200,150)
        
        SURFACE.blit(tmp_msg,tmp_rect)



        pygame.display.update()
        FPSCLOCK.tick(10)

if __name__ =='__main__':
    main()
```

> 마우스 클릭    MOUSEBUTTONDOWN

```python
#pos는 중심좌표   radius는 반경

import sys
import pygame
from pygame.locals import QUIT,MOUSEBUTTONDOWN

pygame.init()
SURFACE = pygame.display.set_mode((400,300))
################
FPSCLOCK = pygame.time.Clock()
#################

def main():
    ### main rutin
    mousepos = []

    while True:
        SURFACE.fill((225,225,225))

        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
            elif event.type == MOUSEBUTTONDOWN:
                mousepos.append(event.pos)

        for pos in mousepos:
            pygame.draw.circle(SURFACE,(0,255,0),pos,5)


        pygame.display.update()
        FPSCLOCK.tick(10)

if __name__ =='__main__':
    main()
```

> 마우스의 이동
- 마우스의 입력은 MOUSEBUTTONDOWN
- 마우스의 이동은 MOUSEMOTION
- 마우스이 해제는 MOUSEBUTTONUP 입니다

```python
#pos는 중심좌표   radius는 반경

import sys
import pygame
from pygame.locals import QUIT,MOUSEBUTTONDOWN,MOUSEMOTION,MOUSEBUTTONUP

pygame.init()
SURFACE = pygame.display.set_mode((400,300))
################
FPSCLOCK = pygame.time.Clock()
#################

def main():
    ### main rutin
    mousepos = []
    mousedown = False

    while True:

        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
            elif event.type == MOUSEBUTTONDOWN:
                mousedown = True
            elif event.type == MOUSEMOTION:
                if mousedown == True:
                    mousepos.append(event.pos)
            elif event.type == MOUSEBUTTONUP:
                mousedown = False
                mousepos=[]

        # SURFACE.fill((225,225,225))
        if len(mousepos) > 1:
            pygame.draw.lines(SURFACE,(255,0,0),False,mousepos,3)

        pygame.display.update()
        FPSCLOCK.tick(10)

if __name__ =='__main__':
    main()
```

> 키누르기  KEYDOWN이벤트로 판별합니다. 키코드는 pygame.locals에 정의 되어있음

```python
#pos는 중심좌표   radius는 반경

import sys
import pygame
from pygame.locals import QUIT,KEYDOWN,K_LEFT,K_RIGHT,K_UP,K_DOWN

pygame.init()
SURFACE = pygame.display.set_mode((800,500))
################
FPSCLOCK = pygame.time.Clock()
#################

pygame.key.set_repeat(5,5)  #거의 동시에 반응할 정도의 속도를 지정한다.

def main():
    ### main rutin
    logo = pygame.image.load("blog.jpg")
    pos = [200,150]
    while True:

        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
            elif event.type == KEYDOWN:
               
                if event.key == K_LEFT:
                    pos[0] -= 2
                elif event.key == K_RIGHT:
                    pos[0] += 2
                elif event.key == K_UP:
                    pos[1] -=5
                elif event.key == K_DOWN:
                    pos[1] += 5
        
        pos[0] = pos[0] % 800
        pos[1] = pos[1] % 500


        SURFACE.fill((225,225,225))
        
        rect= logo.get_rect()
        rect.center = pos
        SURFACE.blit(logo,rect)

        pygame.display.update()
        FPSCLOCK.tick(30)

if __name__ =='__main__':
    main()
```
