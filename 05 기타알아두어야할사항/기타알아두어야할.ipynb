> sin/cos
- 여기에서는 복잡한 거 빼고 단순히 간단히 방향과 크기를 가진 값을 x축,y축 성분으로 변환만 하면 된다

```python

import sys
import pygame
import random
from math import sin,cos,radians
from pygame.locals import QUIT
import time
pygame.init()
SURFACE = pygame.display.set_mode((800,500))
FPSCLOCK = pygame.time.Clock()

def main():
    

    ### main rutin
    while True:

        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
    
        SURFACE.fill((0,0,0))
        # a=random.randint(10,150)
        # b=random.randint(10,150)
        # c=random.randint(10,150)
        # d=random.randint(10,150)
        pointlist0,pointlist1=[],[]
        for theta in range(0,360,120):
            rad= radians(theta)
            pointlist0.append((cos(rad)*100+100,sin(rad)*100+150))
            
        pygame.draw.aalines(SURFACE,(255,255,255),True,pointlist0,5)


        pygame.display.update()
        FPSCLOCK.tick(30)

if __name__ =='__main__':
    main()
```

> 객체지향
- 물건이라는 개념을 프로그래밍 세계에 적용하는 것이 객체지향입니다.
- 일상에서의 물건은  "객체" 또는 "인스턴스"라고 부른다
- 물건의 특징을  "프로퍼티"라고 부른다
- 물건의 조작을  "메서드"라고 부른다
- 객체가 있어야 프로퍼티도 메서드도 있다.
    - 그래서 메서드나 프로퍼티에 접근하는 경우는 다음과 같이 기술한다. 객체변수가 아니면 접근도 호출도 할 수없다   
    - 프로퍼티로의 접근 :  객체변수.프로퍼티명
    - 메서드 호출 : 객체변수.메서드명()
    - 일반적인 함수는 함수명()로 객체 변수가 없어도 실행할 수있다.

```python
# 리스트도 객체입니다.
data = [1,2,3]
data.append(4)
data= [1,2,3,4]  #append라는 메서드를 사용한 것으로 리스트는 객체임을 알 수 있다. data가 객체변수 즉 인스턴스입니다.

#이에 반해
del data[2]
data = [1,2,4]  # del로 소거하는 대상은 리스트에 한정되지 않습니다. 리스트 이외의 요소도 지정할 수 있다. 일반적인 함수에 가가운 사용법을 할 수 있다.
```

> 클래스 는 붕어빵 틀   
> 인스턴스는 붕어빵 들
- 모든 객체는 고유합니다. 같을 수 가 없습니다. 그러나 어떤 종류의 객체들은 같은 특징과 기능을 하고 있습니다. 
- 각각의 객체를 추상화 한 것을 "클래스"라고 합니다. 클래스는 실체가 아닌 추상적인 개념입니다.

> 객체를 만들기 위한 전용함수를 "컨스트럭터(생성자)라고 합니다"
    - 붕어빵을 굽는 쇠로만든 기계를 생각하면 됩니다.

> 붕어빵속에 팥이나 크림이나 초코가 들어있는 붕어빵에는 공통의 특징이 있습니다. 이 특징으로 인하여 다른 과자와 구별할 수 있습니다. 이 공통의 특징이 클래스에 해당합니다. 반복하건데 클래스는 실체가 아닙니다.

- 클래스 설계에 절대적인 정답은 없습니다. 경험이 말하는 세계이므로 여러가지 코드를 읽거나 실패를 거듭하면서 셩험을 쌓아가는 수밖에는 없습니다.   
- 다음과 같은 접근법을 취한다.   
    - 명사(게임 중에는 움직이는 것)는 클래스로 할 수 있는 것이 많다.
    - 비슷한 특징이 있는 것은 같은 클래스 또는 상속관계를 만들 수 있는지 생각한다.
    - 반드시 클래스로 하면 좋은 것은 아니다. 함수가 간단할 수 있다.
    - 상속을 무리하게 사용하는 것보다 클래스를 조합하도록 한다.   
- 그중에서 특히 뛰어난 설계를 카달로그로서 합한것을 [디자인패턴]이라고 하며 이미 많은 디자인 패턴이 서적으로 출판돼있고, 인터넷에도 많은 정보가 공개되어 있다.

> 상속
- 영어에는 reinventing the wheel이라는 말이 있습니다. "이미 발명되어 있는 것을 모르고 다시 발명하는 것은 낭비"라는 의미입니다. 재활용을 중시하는 프로그래밍 업계에서 자주 사용합니다.

```python
import pygame
class MyRect(pygame.Rect):
    def flip(self):
        self.width,self.height = (self.height,self.width)

r=MyRect(10,20,30,40)
r.size
r.center
r.centerx
r.flip()
r.size
r.bottom
r.bottomleft
r.bottomright
```

```python
r= MyRect(10,20,30,40)
r.size
r.flip()
r.size
```

> 다른 클래스를 상속받을때는   
> class  클래스명(부모클래스)

```python
class Audio:
    def __init__(self,power,volume):  #프로퍼티(특성)
        self.power = power
        self.volume = volume

    def switch(self,on_off):  #전원(스위치)조작 메서드
        self.power = on_off

    def set_volume(self,vol):  # 볼륨조절 메서드
        self.volume= vol

    def tune(self): #방송시작 메서드
        str="la la la ..." if self.power else "tune it on"
        print(str)
```

```python
mp3 = Audio(False,8)
mp3.power
mp3.volume
mp3.set_volume = 12
mp3.tune()
mp3.switch(True)
mp3.tune()
```
