04 SNAKE
> 상하좌우 키로 뱀을 조작하여 먹이를먹입니다. 먹이를 하나 먹을 때마나 한 단계 성장합니다. 상하좌우의 벽, 자기에게 부딪히지 않고 어디까지 성장할 수 있을까요?

> 클래스로 구현안 버전을 만들어 봅시다

```python
######### s na ke  ############

import sys
import random
import pygame
from pygame.locals import KEYDOWN, K_UP, K_DOWN, K_LEFT, K_RIGHT, QUIT, Rect

pygame.init()
pygame.key.set_repeat(5,5)
SURFACE = pygame.display.set_mode((600, 600))
FPSCLOCK = pygame.time.Clock()


###################################################################################

class Snake:
    def __init__(self,pos): #스네이크객체생성자
        self.bodies = [pos]  #스네이크몸통좌표목록을 pos로 초기화한다.

    def move(self,key): #스네이크를 1프레임만큼이동
        xpos,ypos = self.bodies[0] #스네이크머리좌표#맨앞의좌표를xpos,ypos로 받는다.
        if key == K_UP:
            ypos -= 1
        elif key == K_DOWN:
            ypos += 1
        elif key == K_LEFT:
            xpos -= 1
        elif key == K_RIGHT:
            xpos += 1
        head=(xpos,ypos)    #키값에 따라 변화된 좌표를 head로 초기화한다.
        #게임오버판정
        is_game_over = head in self.bodies or head[0]<0 or head[0]>=W or head[1]<0 or head[1]>=H

        self.bodies.insert(0, head) #머리좌표를몸통목록의맨앞에추가한다.#   
        if head in FOODS:   #먹이를 먹으면 FOODS에서 먹이를 이동하는 로직만 처리하고 몸통목록에서 꼬리를 제거하지 않는다.이때 bodies목록의길이가1증가한다.
            #먹이를다른장소로이동
            i = FOODS.index(head)
            del FOODS[i]
            add_food(self)
        else:
            self.bodies.pop()

        return is_game_over


    def draw(self):  #스네이크를 그린다
        for body in self.bodies:
            pygame.draw.rect(SURFACE, (0,255,0), Rect(body[0]*30, body[1]*30, 30, 30))
        # pygame.draw.rect(SURFACE, (255,255,0), Rect(self.bodies[0][0]*30, self.bodies[0][1]*30, 30, 30))

FOODS= []
(W,H)=(20,20)

def add_food(snake): #임의의장소에 먹이를 배치
    while True:
        pos = (random.randint(0,W-1), random.randint(0,H-1))
        if pos  in FOODS or pos in snake.bodies:
            continue
        FOODS.append(pos)
        break

def paint(snake,message): #화면 전체를 그린다
    SURFACE.fill((0,0,0))
    snake.draw()
    for food in FOODS:
       pygame.draw.ellipse(SURFACE, (255,0,0), Rect(food[0]*30, food[1]*30, 30, 30))

    for index in range(20):
        pygame.draw.line(SURFACE, (64,64,64), (0,index*30), (600,index*30))
        pygame.draw.line(SURFACE, (64,64,64), (index*30,0), (index*30,600))
    if message != None:
        SURFACE.blit(message, (200,250))
    pygame.display.update()

def main():
    myfont = pygame.font.SysFont(None, 80)
    key = K_DOWN
    message = None
    game_over = False
    snake = Snake((int(W/2),int(H/2)))

    for _ in range(10):
        add_food(snake)
    while True:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
            elif event.type == KEYDOWN:
                key = event.key
        
        if game_over:
            message = myfont.render("Game Over", True, (255,0,0))
        else:
            game_over = snake.move(key)
            pass

        paint(snake, message)
        FPSCLOCK.tick(5)

if __name__ == '__main__':
    main()
        
```
