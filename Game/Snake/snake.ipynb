```python
# 04 Snake  상하좌우 키로 뱀을 조작해서 먹이를 먹입니다. 먹이를 하나 먹을 때마다 한 단계 성장합니다. 상하좌위의 벽, 자기에게 부딪히지 않고 어디까지 성장할 수 있을 까요? 간단한 거에 비해서 중독성 있는 게임이라고 생각합니다.

#소스코드 : 함수버전  (함수만으로 구현함)
#운석 무리를 피해서 어디까지 진행하는지를 겨루는 게임입니다. 모든 이미지는 2D로 사용하고 있습니다.



import sys
import pygame
import time
import random
from pygame.locals import QUIT,KEYDOWN,KEYUP,K_LEFT,K_RIGHT,K_UP,K_DOWN,Rect


pygame.init()
SURFACE = pygame.display.set_mode((600,600))
FPSCLOCK = pygame.time.Clock()
# pygame.mixer.music.load("Escape.mp3")
# pygame.mixer.music.play()
# 종료 pygame.mixer.music.stop()
# hit_sound = pygame.mixer.Sound(r"C:\KCS\Desktop\게임으로배우는 파이썬\Game\1_sounds\bang.wav")   # 짧은 wav/mp3 파일
# eat_sound = pygame.mixer.Sound(r"C:\KCS\Desktop\게임으로배우는 파이썬\Game\1_sounds\eat.wav")   # 짧은 wav/mp3 파일
FOODS = []
SNAKE=[]
(W,H) = (20,20)

def add_food():
    #임의의 장소에 먹이를 배치
    while True:
        pos = (random.randint(0,W-1)), random.randint(0,H-1)
        if pos in FOODS or pos in SNAKE:
            continue

        FOODS.append(pos)
        break

def move_food(pos):
    #먹이를 다른 장소로 이동
    i = FOODS.index(pos)
    del FOODS[i]
    add_food()

def paint(message):
    #화면전체그리기
    SURFACE.fill((0,0,0))

    for food in FOODS:
        pygame.draw.ellipse(SURFACE,(0,255,0),Rect(food[0]*30,food[1]*30,30,30))

    for body in SNAKE:
        if body ==SNAKE[0] :
            pygame.draw.rect(SURFACE,(255,0,0),Rect(body[0]*30,body[1]*30,30,30))
        else:
            pygame.draw.rect(SURFACE,(0,255,255),Rect(body[0]*30,body[1]*30,30,30))
            # pygame.draw.ellipse(SURFACE,(0,255,255),Rect(body[0]*30,body[1]*30,30,30))

    for i in range(20):
        pygame.draw.line(SURFACE,(64,64,64),(i*30,0),(i*30,600))
        pygame.draw.line(SURFACE,(64,64,64),(0,i*30),(600,i*30))

    if message != None:
        SURFACE.blit(message,(150,300))
    pygame.display.update()

def main():
    myfont= pygame.font.SysFont(None,80)
    SNAKE.append((int(W/2),int(H/2)))
    key = K_DOWN
    game_over = False
    for _ in range(10):
        add_food()

    message= None
    while True:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
            elif event.type == KEYDOWN:
                key= event.key
            
        if not game_over:
            if key == K_LEFT:
                head= (SNAKE[0][0]-1,SNAKE[0][1])
            elif key == K_RIGHT:
                head = (SNAKE[0][0]+1,SNAKE[0][1])
            elif key == K_UP:
                head = (SNAKE[0][0],SNAKE[0][1]-1)
            elif key == K_DOWN:
                head = (SNAKE[0][0],SNAKE[0][1]+1)

            if head in SNAKE or head[0] < 0 or head[0] >= W or head[1] < 0 or head[1] >= H:
                message = myfont.render("Game Over!",True,(255,255,0))
                game_over = True
                # hit_sound.play()
                pygame.mixer.music.stop()    

            
            SNAKE.insert(0,head)

            if head in FOODS:
                move_food(head)
                # eat_sound.play()
            else:
                SNAKE.pop()
        paint(message)
        FPSCLOCK.tick(5)

if __name__ =='__main__':
    main()
```

```python
# 04 Snake  상하좌우 키로 뱀을 조작해서 먹이를 먹입니다. 먹이를 하나 먹을 때마다 한 단계 성장합니다. 상하좌위의 벽, 자기에게 부딪히지 않고 어디까지 성장할 수 있을 까요? 간단한 거에 비해서 중독성 있는 게임이라고 생각합니다.

#소스코드 : 객체지향버전  (클래스사용으 구현함)
#



import sys
import pygame
import time
import random
from pygame.locals import QUIT,KEYDOWN,K_LEFT,K_RIGHT,K_UP,K_DOWN,Rect


pygame.init()
pygame.key.set_repeat(5,5)
SURFACE = pygame.display.set_mode((600,600))
FPSCLOCK = pygame.time.Clock()
pygame.mixer.music.load("Escape.mp3")
# pygame.mixer.music.play()
# 종료 pygame.mixer.music.stop()
hit_sound = pygame.mixer.Sound("bang.wav")   # 짧은 wav/mp3 파일
eat_sound = pygame.mixer.Sound("eat.wav")   # 짧은 wav/mp3 파일
# 실행 hit_sound.play()


class Snake:
    #Snake객체
    def __init__(self,pos):
        self.bodies = [pos]
    
    def move(self,key):
        #snake를 1프레임 만큼 이동
        xpos,ypos = self.bodies[0]
        if key == K_LEFT:
            xpos -= 1
        elif key== K_RIGHT:
            xpos += 1
        elif key== K_UP:
            ypos -= 1
        elif key == K_DOWN:
            ypos += 1

        head=(xpos,ypos)

        #게임오버판정
        is_game_over = head in self.bodies or head[0] < 0 or head[0] >=W or head[1] < 0 or head[1] >= H

        self.bodies.insert(0,head)

        if head in FOODS:
            #먹이를 다른 장소로 이동
            i = FOODS.index(head)
            del FOODS[i]
            add_food(self)
            eat_sound.play()
        else:
            self.bodies.pop()

        return is_game_over

    def draw(self):
        #snake를 그린다
        for body in self.bodies:
            pygame.draw.rect(SURFACE,(0,255,255),Rect(body[0]*30,body[1]*30,30,30))

FOODS= []
(W,H) = (20,20)

def add_food(snake):
    #임의의 장소에 먹이를 배치
    while True:
        pos = (random.randint(0,W-1)), random.randint(0,H-1)
        if pos in FOODS or pos in snake.bodies:
            continue
        FOODS.append(pos)
        break

def paint(snake,message):
    #화면전체그리기
    SURFACE.fill((0,0,0))
    snake.draw()

    for food in FOODS:
        pygame.draw.ellipse(SURFACE,(0,255,0),Rect(food[0]*30,food[1]*30,30,30))

    # for body in SNAKE:
    #     if body ==SNAKE[0] :
    #         pygame.draw.rect(SURFACE,(255,0,0),Rect(body[0]*30,body[1]*30,30,30))
    #     else:
    #         pygame.draw.rect(SURFACE,(0,255,255),Rect(body[0]*30,body[1]*30,30,30))
    #         # pygame.draw.ellipse(SURFACE,(0,255,255),Rect(body[0]*30,body[1]*30,30,30))

    for i in range(20):
        pygame.draw.line(SURFACE,(64,64,64),(i*30,0),(i*30,600))
        pygame.draw.line(SURFACE,(64,64,64),(0,i*30),(600,i*30))

    if message != None:
        SURFACE.blit(message,(150,300))
    pygame.display.update()

def main():
    myfont= pygame.font.SysFont(None,80)
    key = K_DOWN
    message= None
    game_over = False
    snake=Snake((int(W/2),int(H/2)))
    for _ in range(10):
        add_food(snake)

    while True:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
            elif event.type == KEYDOWN:
                key= event.key
            
        if game_over:
            message=myfont.render("Game Over!!",True,(255,255,0))
            pygame.mixer.music.stop()
            hit_sound.play()
            time.sleep(3)
            pygame.quit()
            sys.exit()
        else: 
            game_over = snake.move(key)

        paint(snake,message)
        FPSCLOCK.tick(5)

if __name__ =='__main__':
    main()
```
