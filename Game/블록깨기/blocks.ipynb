```python
# 05 블록깨기    설명이 필요 없는 유명한 게임입니다. 화면 아래에 있는 패들을 조작하여 공을 튕겨서 화면 위에 있는 블록을 전부 제거하면 됩니다.

import sys
import math
import pygame
import time
import random
from pygame.locals import QUIT,KEYDOWN,K_LEFT,K_RIGHT,Rect

class Block:
    ##블록,공,패들 객체##
    def __init__(self,col,rect,speed=0):
        self.col = col
        self.rect = rect
        self.speed = speed
        self.dir = random.randint(-45,45) -270

    def move(self):
        # 공을 움직인다
        self.rect.centerx +=math.cos(math.radians(self.dir))*self.speed
        self.rect.centery -=math.sin(math.radians(self.dir))*self.speed

    def draw(self):
        ##블록,공,패들을 그린다
        if self.speed == 0:
            pygame.draw.rect(SURFACE,self.col,self.rect)
        else:
            pygame.draw.ellipse(SURFACE,self.col,self.rect)

def tick():
    # 프레임별처리
    global BLOCKS

    for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
            elif event.type == KEYDOWN:
                if event.key == K_LEFT:
                    if PADDLE.rect.left > 0 : #좌측벗어나는거 제한
                        PADDLE.rect.centerx -=1  
                elif event.key == K_RIGHT:
                    if PADDLE.rect.right < 600: #우측벗어나는거 제한
                        PADDLE.rect.centerx += 1  

    if BALL.rect.centery < 1000:
        BALL.move()    

    # 패들과 충돌 ##
    if PADDLE.rect.colliderect(BALL.rect):
        BALL.dir = 90 + (PADDLE.rect.centerx - BALL.rect.centerx) / PADDLE.rect.width*80
        eat_sound.play()
    # 블록과 충돌
    prevlen = len(BLOCKS)

    BLOCKS = [x for x in BLOCKS if not x.rect.colliderect(BALL.rect)]
        #BLOCKS 안에 있는 블록들 중에서 공(BALL)과 부딪히지 않은 블록만 남겨둬.즉, 부딪힌 블록은 리스트에서 제거돼.
    if len(BLOCKS) != prevlen:
        BALL.dir *= -1  #(공의 이동 방향)을 약간 바꿔서 튕기게 만들어.
        eat_sound.play()

    
    # 벽과충돌
    if BALL.rect.centerx <0 or BALL.rect.centerx >600 :
        BALL.dir = 180 - BALL.dir
        eat_sound.play()
    if BALL.rect.centery  <0:
        BALL.dir = -BALL.dir
        BALL.speed= 15
        eat_sound.play()

pygame.init()
pygame.key.set_repeat(1,1)
SURFACE = pygame.display.set_mode((600,800))
FPSCLOCK = pygame.time.Clock()
# pygame.mixer.music.load("Escape.mp3")
# pygame.mixer.music.play()
# 종료 pygame.mixer.music.stop()
# hit_sound = pygame.mixer.Sound("bang.wav")   # 짧은 wav/mp3 파일
eat_sound = pygame.mixer.Sound("eat.wav")   # 짧은 wav/mp3 파일
# 실행 hit_sound.play()


####시작준비화면####################################################################################
pygame.display.set_caption("블록깨기")
# 글꼴 설정
font = pygame.font.SysFont(None, 36)
text = font.render("Start press SPACE!", True, (100,100,100))
text_rect = text.get_rect(center=(600//2,800//2))

# --- 1단계: 대기 화면 ---
waiting = True
while waiting:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            raise SystemExit
        if event.type == pygame.KEYDOWN and event.key == pygame.K_SPACE:
            waiting = False  # 스페이스키 눌리면 대기 종료
    SURFACE.fill((0,0,0))
    SURFACE.blit(text, text_rect)
    pygame.display.flip()
    FPSCLOCK.tick(30)
###################################################################################

BLOCKS= []
PADDLE= Block((242,242,0),Rect(300,700,100,30))
BALL = Block((242,242,0),Rect(300,400,20,20),10)

def main():
    myfont= pygame.font.SysFont(None,80)
    remainfont= pygame.font.SysFont(None,32)

    mess_clear = myfont.render("Cleared !!",True,(255,255,0))
    mess_over = myfont.render("Game Over !!",True,(255,255,0))
    fps = 30
    colors =[(255,0,0),(255,160,0),(240,240,0),(0,128,0),(128,0,128),(0,0,250)]


    for ypos, color in enumerate(colors,start=0):
        for xpos in range(0,5):
            BLOCKS.append(Block(color,Rect(xpos*100 + 60, ypos* 40 + 40, 80,30)))
        #, enumerate()는 내부적으로 (index, value) 형태의 **튜플을 반환하는 반복자(iterator)**야.시작값지정도가능 start=1
    while True:
        
        tick()
        SURFACE.fill((0,0,0))
        prevlen = len(BLOCKS)

        remain_str= str(prevlen).zfill(2)  #즉 zfill(6)은 전체 길이를 6자리로 맞추겠다는 뜻이고, 부족한 자리만큼 왼쪽에 0을 채워 넣는 거지.
        remain_image = remainfont.render(remain_str,True,(0,255,0))
        SURFACE.blit(remain_image,(50,600))
        BALL.draw()
        PADDLE.draw()
        
        for block in BLOCKS:
            block.draw()

        if len(BLOCKS) == 0 :
            SURFACE.blit(mess_clear,(200,400))
        if BALL.rect.centery > 800 and len(BLOCKS) >0 :
            SURFACE.blit(mess_over,(150,400))
            
        pygame.display.update()
        FPSCLOCK.tick(fps)
        

if __name__ =='__main__':
    main()
```
