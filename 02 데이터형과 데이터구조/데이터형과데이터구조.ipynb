map은  리스트나 튜플의 요소 전체에 대하여 어떠한 처리를 할 때 사용합니다


```python
def make_double(x):
    return x*2

make_double(3)

list(map(make_double,[2,3,4]))

# lambda식을 이용하여 간단하게 아래와 같이 함수없이 표현 할 수 있다.
list(map(lambda x:x*2,[1,2,3]))
```
